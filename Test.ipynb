```python
urls = []
platform = ['fd','dk','yh']
for plat in platform:
    for year in range(17,22):
        if year <=20:
            for week in range (1,18):
                url = f'http://rotoguru1.com/cgi-bin/fyday.pl?week={week}&year=20{year}&game={plat}&scsv=1'
                urls.append(url)
        else:
            for week in range (1,6):
                url = f'http://rotoguru1.com/cgi-bin/fyday.pl?week={week}&year=20{year}&game={plat}&scsv=1'
                urls.append(url)
print(urls)
```

```python
def col_str_to_float(df, col):
    df[col] = [val.replace('.', '0').strip() if val == '.' else val for val in df[col]]
    df[col]  = [val.replace('', '0').strip() if val == '' else val for val in df[col]]
    df[col]  = [val.replace('. 0', '0').strip() if val == '. 0' else val for val in df[col]]


    df[col] = df[col].apply(lambda s : str(s).replace('.-', '-.').strip())
    # = [(val.replace('.-', '-.').strip()) for val in data_2020['FDP']]
    df[col] = [float(val) for val in df[col]]
```
